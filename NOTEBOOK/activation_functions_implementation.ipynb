```python
# NAME : KAUSHAL SHETE
# PRN : 20220802168
# BATCH : DS-1

import numpy as np
def step_function(x):
    return 1 if x>=0 else 0

def perceptron(inputs, weights, bias):
    # calculate weighted sum
    weighted_sum = np.dot(inputs, weights) + bias
    return step_function(weighted_sum)
```

```python
# AND Gate
inputs_and = np.array([[0,0], [0,1], [1,0], [1,1]])
outputs_and = np.array([0,0,0,1])
weights_and = np.array([1,1])
bias_and = -1.5

print("AND Gate:")
for i in range(len(inputs_and)):
    predicted_output = perceptron(inputs_and[i], weights_and, bias_and)
    print(f"Input: {inputs_and[i]}, Predicted Output: {predicted_output}")
```

```python
# OR Gate
inputs_or = np.array([[0,0], [0,1], [1,0], [1,1]])
outputs_or = np.array([0,1,1,1])
weights_or = np.array([1,1])
bias_or = -0.5

print("\nOR Gate:")
for i in range(len(inputs_or)):
    predicted_output = perceptron(inputs_or[i], weights_or, bias_or)
    print(f"Input: {inputs_or[i]}, Predicted Output: {predicted_output}")
```

```python
# NOT Gate
inputs_nor = np.array([[0,0], [0,1], [1,0], [1,1]])
outputs_nor = np.array([1,0,0,0])
weights_nor = np.array([-1,-1])
bias_nor = 0.5

print("\nNOT Gate:")
for i in range(len(inputs_nor)):
    predicted_output = perceptron(inputs_nor[i], weights_nor, bias_nor)
    print(f"Input: {inputs_nor[i]}, Predicted Output: {predicted_output}")
```

```python
# XOR Gate
inputs_xor = np.array([[0,0], [0,1], [1,0], [1,1]])
outputs_xor = np.array([0,1,1,0])
weights_xor = np.array([1,1])
bias_xor = -1

print("\nXOR Gate:")
for i in range(len(inputs_xor)):
    predicted_output = perceptron(inputs_xor[i], weights_xor, bias_xor)
    print(f"Input: {inputs_xor[i]}, Predicted Output: {predicted_output}")
```

```python
# XNOR Gate
inputs_xnor = np.array([[0,0], [0,1], [1,0], [1,1]])
outputs_xnor = np.array([0,1,1,0])
weights_xnor = np.array([1,1])
bias_xnor = -1

print("\nXNOR Gate:")
for i in range(len(inputs_xnor)):
    predicted_output = perceptron(inputs_xnor[i], weights_xnor, bias_xnor)
    print(f"Input: {inputs_xnor[i]}, Predicted Output: {predicted_output}")
```
