```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
coefficients = np.polyfit(x,y,2)
```

```python
poly = np.poly1d(coefficients)
```

```python
plt.scatter(x,y, color = 'blue', label = 'Data Points')
```

```python
x_line = np.linspace(min(x), max(x), 100)
```

```python
plt.plot(x_line,poly(x_line), color='red', label='Fitted polynomial')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.legend()

plt.show()
```

```python
new_x = 6
predicted_y = poly(new_x)
print(f"Predicted y value for x {new_x}: {predicted_y}")
```
