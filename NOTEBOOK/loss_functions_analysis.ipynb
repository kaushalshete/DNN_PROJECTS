```python
# NAME : KAUSHAL SHETE
# PRN : 20220802168
# BATCH : DS-1

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

def two_layer_perceptron(X, W1, b1, W2, b2):
    return softmax(np.dot(sigmoid(np.dot(X, W1) + b1), W2) + b2)

activation_functions = {
    "Sigmoid": sigmoid,
    "Tanh": np.tanh,
    "ReLU": lambda x: np.maximum(0, x),
    "Leaky ReLU": lambda x: np.where(x > 0, x, 0.01 * x),
    "ELU": lambda x: np.where(x > 0, x, np.exp(x) - 1),
    "Softplus": lambda x: np.log1p(np.exp(x)),
    "Swish": lambda x: x * sigmoid(x)
}

x = np.linspace(-10, 10, 400)
sample_data = np.array([-2, -1, 0, 1, 2])

for name, func in activation_functions.items():
    plt.figure()
    plt.plot(x, func(x), label=name, color='blue')
    plt.scatter(sample_data, func(sample_data), color='red', label='Sample Points')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title(f"{name} Activation Function")
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.legend()
    plt.grid(True)
    plt.show()

data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
np.random.seed(42)
W1, b1, W2, b2 = np.random.randn(2, 4), np.random.randn(4), np.random.randn(4, 2), np.random.randn(2)
print("Predictions:", two_layer_perceptron(data, W1, b1, W2, b2))
```
